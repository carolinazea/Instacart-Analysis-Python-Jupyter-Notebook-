# If-Statements with User-Defined Functions

```python
# Define a function 

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return np.nan
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#Since there are no high-range products, use the max() function to look up the highest priced product

df['prices'].max()
```

```python
# You know 0 starts from Sunday and 4 is a Thursday. 

#To display all rows under 'orders_day_of_week' column to least, regular and busiest day (use loop)
```

```python
result = []

for x in ords_prods_merged["orders_day_of_week"]:
  if x == (0):
    result.append("Busiest day")
  elif x == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```
