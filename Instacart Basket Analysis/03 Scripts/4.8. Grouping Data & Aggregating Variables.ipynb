## Grouping Data with pandas

```python
# Step 2 

Calculate the mean 'order_number' column grouped by the 'department_id' column (compare avg # of orders across depts)
 #1. Split the data into groups based on a column, 'department_id'
 #2. Apply the agg() function to each group to obrain the mean values for the 'order_number' column 
```

## Two ways on how to find the mean in a column using groupby 

```python
# one way to conduct formula 

df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# another way to conduct the same forumula

df.groupby('department_id')['order_number'].mean()
```

## How to produce multiple statistics (mean, min, max) in a column using groupby

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## Create statistical flags for data using transform( )  [last column:max_order]

## transform( ) create new column containing the statistic (mean, min, max) of column you want to groupby 

```python
## As above, the min is 1 and max is 99 on a continous note, create flag to say 'most ordered', 'loyalty members', etc
```

```python
# In one single code, 
# 1. Split the data into groups based on the “user_id” column.
# 2. Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# 3. Create a new column, “max_order,” into which you’ll place the results of your aggregation.
```

```python
ords_prods_merge['max_order'] = # The new column will be called 'max_order'
ords_prods_merge.groupby(['user_id']) # Column you want to split into groups 
['order_number'].transform(np.max) # Transform uses numpy (np) to find max values in 'order_number' column 
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## Create labeled flags for data using loc( ) based on the 'max_order' column created

## loc ( ) create a second column containing the flag description 

```python
# 3 lines of code into one 
# uses 'max_order' to flag 
# new labeled column that will be created is called 'loyalty_flag'
# the results are either 'loyal customer', 'Regular customer' or 'New customer'

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer' 

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10),
'loyalty_flag'] = 'Regular customer'

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
